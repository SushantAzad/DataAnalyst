### Assignment 3

```python
pos_set1 = set(x for x in range(1,6))
pos_set2 = set(x for x in range(1,11) if x %2==0)
print(pos_set1)
print(pos_set2)

union_set = pos_set1.union(pos_set2)
inters_set = pos_set1.intersection(pos_set2)
diff = pos_set1.difference(pos_set2)
symm_diff = pos_set1.symmetric_difference(pos_set2)
print(union_set)
print(inters_set)
print(diff)
print(symm_diff)
```
