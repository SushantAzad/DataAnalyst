### Assignment 15

```python
def minde(tup):
    return min(tup)
def maxde(tup):
    return max(tup)
def sumde(tup):
    return sum(tup)

tup1 = (1,2,3,4,5,6)
print(minde(tup1))
print(maxde(tup1))
print(sumde(tup1))
```
