# NumPy Functions Cheat Sheet

| Function | Description | Example (Single Line) |
|----------|-------------|------------------------|
| `np.array()` | Creates a NumPy array from a list, tuple, or other iterable. | `arr = np.array([1, 2, 3, 4])` |
| `np.arange()` | Creates an array with evenly spaced values within a given range. | `arr = np.arange(0, 10, 2)` |
| `np.linspace()` | Creates an array with a specified number of equally spaced values between two numbers. | `arr = np.linspace(0, 1, 5)` |
| `np.zeros()` | Creates an array filled with zeros. | `arr = np.zeros((2, 3))` |
| `np.ones()` | Creates an array filled with ones. | `arr = np.ones((3, 2))` |
| `np.full()` | Creates an array filled with a specified value. | `arr = np.full((2, 2), 7)` |
| `np.eye()` | Creates an identity matrix with ones on the main diagonal. | `arr = np.eye(3)` |
| `np.reshape()` | Changes the shape of an array without changing its data. | `arr = np.reshape(np.arange(6), (2, 3))` |
| `np.flatten()` | Converts a multidimensional array into a one dimensional array. | `arr = np.array([[1, 2], [3, 4]]).flatten()` |
| `np.transpose()` | Transposes rows and columns of an array. | `arr = np.transpose(np.array([[1, 2], [3, 4]]))` |
| `np.mean()` | Calculates the average of array elements. | `mean = np.mean([1, 2, 3, 4, 5])` |
| `np.sum()` | Calculates the sum of array elements. | `total = np.sum([1, 2, 3, 4])` |
| `np.max()` | Returns the largest value in an array. | `maximum = np.max([5, 8, 2, 9])` |
| `np.min()` | Returns the smallest value in an array. | `minimum = np.min([5, 8, 2, 9])` |
| `np.argmax()` | Returns the index of the maximum value. | `index = np.argmax([5, 8, 2, 9])` |
| `np.argmin()` | Returns the index of the minimum value. | `index = np.argmin([5, 8, 2, 9])` |
| `np.where()` | Returns indices where a specified condition is true. | `indices = np.where(np.array([1, 2, 3, 4]) > 2)` |
| `np.unique()` | Returns unique values from an array. | `unique_values = np.unique([1, 2, 2, 3, 3, 4])` |
| `np.concatenate()` | Joins two or more arrays along an existing axis. | `arr = np.concatenate((np.array([1, 2]), np.array([3, 4])))` |
| `np.random.rand()` | Generates random numbers between 0 and 1. | `arr = np.random.rand(2, 3)` |
